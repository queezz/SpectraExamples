```python
import pandas as pd
import numpy as np
import os
from os.path import join
```

```python
bpth = './data' # set base folder path
ls = sorted(os.listdir(bpth)) # list all files in it
expnum = '0306_05'
ls = [i for i in ls if expnum in i] # select files with 'expnum' in names
fp = join(bpth,ls[0])
fp1 = join(bpth,ls[1])
fp,fp1
```

```python
data = pd.read_csv(fp,skiprows=66,sep=';',na_values=['   ', ' '], dtype=float)
data1 = pd.read_csv(fp1,skiprows=66,sep=';',na_values=['   ', ' '], dtype=float)
```

```python
font_setup(size=15) # set font size in the plot
gcf().set_size_inches(8,6) # define image size
plot(data['Wavelength'],data['Raw data #1'],label='plasma')
plot(data1['Wavelength'],data1['Raw data #1'],label='filament')

xlabel('wavelength, nm')
ylabel('Intensity, a.u.')
gritix()
legend()
#yscale('log')

figname = f'{expnum}-test.png'
savefig(join('./img',figname),dpi=200,bbox_inches='tight')
```
